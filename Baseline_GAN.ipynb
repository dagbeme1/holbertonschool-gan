```python
import os
import gdown
from zipfile import ZipFile
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random

# Create a directory to store the dataset
try:
    os.makedirs("celeba_gan")
except FileExistsError:
    pass

# Download the CelebA dataset
url = "https://drive.google.com/uc?id=1O7m1010EJjLE5QxLZiM9Fpjs7Oj6e684"
output = "celeba_gan/data.zip"
gdown.download(url, output, quiet=False)

# Extract the dataset
with ZipFile("celeba_gan/data.zip", "r") as zipobj:
    zipobj.extractall("celeba_gan")

# List of image file names in the dataset directory
image_dir = "celeba_gan/img_align_celeba"
image_files = os.listdir(image_dir)

# Select a random image or specify the file name of the image you want to see
# Replace 'your_image.jpg' with the actual file name
image_to_display = random.choice(image_files)  # Random image
# image_to_display = 'your_image.jpg'  # Specific image

# Load and display the selected image
image_path = os.path.join(image_dir, image_to_display)
sample_image = cv2.imread(image_path)[..., ::-1]  # Read and convert BGR to RGB
plt.imshow(sample_image)
plt.title("Sample Image from CelebA Dataset")
plt.axis('off')
plt.show()
```
